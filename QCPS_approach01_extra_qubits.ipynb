# First approach (additional qubits). Three qubits case

## Define functions

```python
import qiskit as qk
import numpy as np
import math as mth
import pandas as pd
import qiskit.tools.jupyter
import matplotlib.pyplot as plt
import seaborn as sns

from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit

"""Function second level"""

def quantum_circuit_three_qubits_configuration_extra(z_a_1,
                                               z_b0_1,
                                               z_bab0_100, z_bab0_101, z_bab0_110, z_bab0_111, 
                                               z_c0_1,
                                               z_cbc0_100, z_cbc0_101, z_cbc0_110, z_cbc0_111):
    qreg_q = QuantumRegister(6, 'q')
    creg_c = ClassicalRegister(6, 'c')
    circuit = QuantumCircuit(qreg_q, creg_c)

# z_a_1 = P(A=|1>)
# z_b0_1 = P(B0=|1>)
# z_c0_1 = P(C0=|1>)

    circuit.reset(qreg_q[0]) # Node C
    circuit.reset(qreg_q[1]) # Ancilla Node
    circuit.reset(qreg_q[2]) # Node B
    circuit.reset(qreg_q[3]) # Node C0
    circuit.reset(qreg_q[4]) # Node A
    circuit.reset(qreg_q[5]) # Node B0
    
    theta_z_a_1=2*mth.atan(np.sqrt(z_a_1/(1-z_a_1)))
    theta_z_b0_1=2*mth.atan(np.sqrt(z_b0_1/(1-z_b0_1)))
    
    theta_z_bab0_100=2*mth.atan(np.sqrt(z_bab0_100/(1-z_bab0_100)))
    theta_z_bab0_101=2*mth.atan(np.sqrt(z_bab0_101/(1-z_bab0_101)))
    theta_z_bab0_110=2*mth.atan(np.sqrt(z_bab0_110/(1-z_bab0_110)))
    theta_z_bab0_111=2*mth.atan(np.sqrt(z_bab0_111/(1-z_bab0_111)))
    
    theta_z_c0_1=2*mth.atan(np.sqrt(z_c0_1/(1-z_c0_1)))
    
    theta_z_cbc0_100=2*mth.atan(np.sqrt(z_cbc0_100/(1-z_cbc0_100)))
    theta_z_cbc0_101=2*mth.atan(np.sqrt(z_cbc0_101/(1-z_cbc0_101)))
    theta_z_cbc0_110=2*mth.atan(np.sqrt(z_cbc0_110/(1-z_cbc0_110)))
    theta_z_cbc0_111=2*mth.atan(np.sqrt(z_cbc0_111/(1-z_cbc0_111)))


    circuit.u(theta_z_bab0_111/2, 0, 0, qreg_q[2])
    circuit.u(theta_z_a_1, 0, 0, qreg_q[4])
    circuit.u(theta_z_b0_1, 0, 0, qreg_q[5])
    circuit.ccx(qreg_q[5], qreg_q[4], qreg_q[1])
    circuit.cx(qreg_q[1], qreg_q[2])
    circuit.u(-theta_z_bab0_111/2, 0, 0, qreg_q[2])
    circuit.cx(qreg_q[1], qreg_q[2])
    circuit.u(theta_z_bab0_100/2, 0, 0, qreg_q[2])
    circuit.ccx(qreg_q[5], qreg_q[4], qreg_q[1])
    circuit.x(qreg_q[4])
    circuit.x(qreg_q[5])
    circuit.ccx(qreg_q[5], qreg_q[4], qreg_q[1])
    circuit.cx(qreg_q[1], qreg_q[2])
    circuit.u(-theta_z_bab0_100/2, 0, 0, qreg_q[2])
    circuit.cx(qreg_q[1], qreg_q[2])
    circuit.u(theta_z_bab0_110/2, 0, 0, qreg_q[2])
    circuit.ccx(qreg_q[5], qreg_q[4], qreg_q[1])
    circuit.x(qreg_q[4])
    circuit.x(qreg_q[5])
    circuit.x(qreg_q[4])
    circuit.ccx(qreg_q[5], qreg_q[4], qreg_q[1])
    circuit.cx(qreg_q[1], qreg_q[2])
    circuit.u(-theta_z_bab0_110/2, 0, 0, qreg_q[2])
    circuit.cx(qreg_q[1], qreg_q[2])
    circuit.u(theta_z_bab0_101/2, 0, 0, qreg_q[2])
    circuit.ccx(qreg_q[5], qreg_q[4], qreg_q[1])
    circuit.x(qreg_q[4])
    circuit.x(qreg_q[5])
    circuit.ccx(qreg_q[5], qreg_q[4], qreg_q[1])
    circuit.cx(qreg_q[1], qreg_q[2])
    circuit.u(-theta_z_bab0_101/2, 0, 0, qreg_q[2])
    circuit.cx(qreg_q[1], qreg_q[2])
    circuit.ccx(qreg_q[5], qreg_q[4], qreg_q[1])
    circuit.x(qreg_q[5])

    circuit.u(theta_z_cbc0_111/2, 0, 0, qreg_q[0])
    circuit.u(theta_z_c0_1, 0, 0, qreg_q[3])
    circuit.ccx(qreg_q[3], qreg_q[2], qreg_q[1])
    circuit.cx(qreg_q[1], qreg_q[0])
    circuit.u(-theta_z_cbc0_111/2, 0, 0, qreg_q[0])
    circuit.cx(qreg_q[1], qreg_q[0])
    circuit.u(theta_z_cbc0_100/2, 0, 0, qreg_q[0])
    circuit.ccx(qreg_q[3], qreg_q[2], qreg_q[1])
    circuit.x(qreg_q[2])
    circuit.x(qreg_q[3])
    circuit.ccx(qreg_q[3], qreg_q[2], qreg_q[1])
    circuit.cx(qreg_q[1], qreg_q[0])
    circuit.u(-theta_z_cbc0_100/2, 0, 0, qreg_q[0])
    circuit.cx(qreg_q[1], qreg_q[0])
    circuit.u(theta_z_cbc0_110/2, 0, 0, qreg_q[0])
    circuit.ccx(qreg_q[3], qreg_q[2], qreg_q[1])
    circuit.x(qreg_q[2])
    circuit.x(qreg_q[3])
    circuit.x(qreg_q[2])
    circuit.ccx(qreg_q[3], qreg_q[2], qreg_q[1])
    circuit.cx(qreg_q[1], qreg_q[0])
    circuit.u(-theta_z_cbc0_110/2, 0, 0, qreg_q[0])
    circuit.cx(qreg_q[1], qreg_q[0])
    circuit.u(theta_z_cbc0_101/2, 0, 0, qreg_q[0])
    circuit.ccx(qreg_q[3], qreg_q[2], qreg_q[1])
    circuit.x(qreg_q[2])
    circuit.x(qreg_q[3])
    circuit.ccx(qreg_q[3], qreg_q[2], qreg_q[1])
    circuit.cx(qreg_q[1], qreg_q[0])
    circuit.u(-theta_z_cbc0_101/2, 0, 0, qreg_q[0])
    circuit.cx(qreg_q[1], qreg_q[0])
    circuit.ccx(qreg_q[3], qreg_q[2], qreg_q[1])
    circuit.x(qreg_q[3])

    circuit.measure(qreg_q[5], creg_c[5])
    circuit.measure(qreg_q[4], creg_c[4])
    circuit.measure(qreg_q[3], creg_c[3])
    circuit.measure(qreg_q[2], creg_c[2])
    circuit.measure(qreg_q[0], creg_c[0])

    # Using Qiskit Aer's Qasm Simulator
    simulator = qk.BasicAer.get_backend('qasm_simulator')

    # Simulating the circuit using the simulator to get the result
    job = qk.execute(circuit, simulator)
    result = job.result()

    # Getting the aggregated binary outcomes of the circuit.
    counts = result.get_counts(circuit)

    # Probability of qc is |0> is
    if '000000' in counts: 
        c_1=counts.get('000000')/1024
    else:
        c_1=0
    if '000010' in counts: 
        c_2=counts.get('000010')/1024
    else:
        c_2=0
    if '000110' in counts: 
        c_3=counts.get('000110')/1024
    else:
        c_3=0        
    if '001110' in counts: 
        c_4=counts.get('001110')/1024
    else:
        c_4=0
    if '011110' in counts: 
        c_5=counts.get('011110')/1024
    else:
        c_5=0
    if '000100' in counts: 
        c_6=counts.get('000100')/1024
    else:
        c_6=0
    if '001100' in counts: 
        c_7=counts.get('001100')/1024
    else:
        c_7=0        
    if '011100' in counts: 
        c_8=counts.get('011100')/1024
    else:
        c_8=0
    if '001000' in counts: 
        c_9=counts.get('001000')/1024
    else:
        c_9=0
    if '011000' in counts: 
        c_10=counts.get('011000')/1024
    else:
        c_10=0
    if '010010' in counts: 
        c_11=counts.get('010010')/1024
    else:
        c_11=0
    if '010000' in counts: 
        c_12=counts.get('010000')/1024
    else:
        c_12=0
    if '010100' in counts: 
        c_13=counts.get('010100')/1024
    else:
        c_13=0
    if '011010' in counts: 
        c_14=counts.get('011010')/1024
    else:
        c_14=0
    if '010110' in counts: 
        c_15=counts.get('010110')/1024
    else:
        c_15=0
    if '001010' in counts: 
        c_16=counts.get('001010')/1024
    else:
        c_16=0
        
    if '100000' in counts: 
        c_17=counts.get('100000')/1024
    else:
        c_17=0
    if '100010' in counts: 
        c_18=counts.get('100010')/1024
    else:
        c_18=0
    if '100110' in counts: 
        c_19=counts.get('100110')/1024
    else:
        c_19=0        
    if '101110' in counts: 
        c_20=counts.get('101110')/1024
    else:
        c_20=0
    if '111110' in counts: 
        c_21=counts.get('111110')/1024
    else:
        c_21=0
    if '100100' in counts: 
        c_22=counts.get('100100')/1024
    else:
        c_22=0
    if '101100' in counts: 
        c_23=counts.get('101100')/1024
    else:
        c_23=0        
    if '111100' in counts: 
        c_24=counts.get('111100')/1024
    else:
        c_24=0
    if '101000' in counts: 
        c_25=counts.get('101000')/1024
    else:
        c_25=0
    if '111000' in counts: 
        c_26=counts.get('111000')/1024
    else:
        c_26=0
    if '110010' in counts: 
        c_27=counts.get('110010')/1024
    else:
        c_27=0
    if '110000' in counts: 
        c_28=counts.get('110000')/1024
    else:
        c_28=0
    if '110100' in counts: 
        c_29=counts.get('110100')/1024
    else:
        c_29=0
    if '111010' in counts: 
        c_30=counts.get('111010')/1024
    else:
        c_30=0
    if '110110' in counts: 
        c_31=counts.get('110110')/1024
    else:
        c_31=0
    if '101010' in counts: 
        c_32=counts.get('101010')/1024
    else:
        c_32=0

    q0_level1_circuit_is_0=(c_1)+(c_2)+(c_3)+(c_4)+(c_5)+(c_6)+(c_7)+(c_8)+(c_9)+(c_10)+(c_11)+(c_12)+(c_13)+(c_14)+(c_15)+(c_16)+(c_17)+(c_18)+(c_19)+(c_20)+(c_21)+(c_22)+(c_23)+(c_24)+(c_25)+(c_26)+(c_27)+(c_28)+(c_29)+(c_30)+(c_31)+(c_32)

    return q0_level1_circuit_is_0
```
